# Articles - Pipeline Notebook

Dataset: `articles.csv`

## 0 - Setup: SparkSession

```python

import sys
sys.path.insert(0, '/app')  # Docker working directory

from pathlib import Path
from IPython.display import HTML, display

from src.articles.notebook_display import spark_df_to_scrollable_html

PROJECT_ROOT = Path('/app')

from src.spark_utils import create_spark_session

spark = create_spark_session('Articles-Notebook')
print('SparkSession ready. Version:', spark.version)
```

## Stage 3 - Extraction

```python
from src.articles.extraction import run_extraction

raw_df = run_extraction(spark, raw_dir=str(PROJECT_ROOT / 'data' / 'raw'))
display(HTML(spark_df_to_scrollable_html(raw_df, rows=5)))
```

## Stage 4 - Preprocessing

```python
from src.articles.preprocessing import run_preprocessing

clean_df = run_preprocessing(raw_df)
print('Rows after preprocessing:', clean_df.count())
display(HTML(spark_df_to_scrollable_html(clean_df, rows=5)))
```

## Save Cleaned Data

```python
OUT_PATH = str(PROJECT_ROOT / 'data' / 'processed' / 'articles')
clean_df.write.mode('overwrite').parquet(OUT_PATH)
print(f'Saved to: {OUT_PATH}')
display(HTML(spark_df_to_scrollable_html(spark.read.parquet(OUT_PATH), rows=5)))
```

```python
spark.stop()
print('SparkSession stopped.')
```
